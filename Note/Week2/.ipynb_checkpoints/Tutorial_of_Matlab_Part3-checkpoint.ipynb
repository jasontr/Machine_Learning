# Plotting Data

```matlab
t = [0: 0.01: 0.98];
```

```matlab
y1 = sin(2 * pi * 4 * t);
```

```matlab
plot(t, y1)
```

```matlab
y2 = cos(2 * pi * 4 * t);
```

```matlab
plot(t, y2)
```

```matlab
plot(t, y1);
hold on;
plot(t, y2, 'r');
xlabel('time');
ylabel('value');
legend('sin', 'cos');
title('my plot');
print -dpng 'myPlot.png'
```

```matlab
close
```

```matlab
figure(1); plot(t, y1);
figure(2); plot(t, y2);
```

```matlab
subplot(1, 2, 1); % Divides plot a 1x2 grid, access first element
plot(t, y1);
subplot(1, 2, 2);
plot(t, y2);
axis([0.5 1 -1 1]);
```
